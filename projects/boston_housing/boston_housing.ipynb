### Implementation: Define a Performance Metric
It is difficult to measure the quality of a given model without quantifying its performance over training and testing. This is typically done using some type of performance metric, whether it is through calculating some type of error, the goodness of fit, or some other useful measurement. For this project, you will be calculating the [*coefficient of determination*](http://stattrek.com/statistics/dictionary.aspx?definition=coefficient_of_determination), R<sup>2</sup>, to quantify your model's performance. The coefficient of determination for a model is a useful statistic in regression analysis, as it often describes how "good" that model is at making predictions. 

The values for R<sup>2</sup> range from 0 to 1, which captures the percentage of squared correlation between the predicted and actual values of the **target variable**. A model with an R<sup>2</sup> of 0 is no better than a model that always predicts the *mean* of the target variable, whereas a model with an R<sup>2</sup> of 1 perfectly predicts the target variable. Any value between 0 and 1 indicates what percentage of the target variable, using this model, can be explained by the **features**. _A model can be given a negative R<sup>2</sup> as well, which indicates that the model is **arbitrarily worse** than one that always predicts the mean of the target variable._

For the `performance_metric` function in the code cell below, you will need to implement the following:
- Use `r2_score` from `sklearn.metrics` to perform a performance calculation between `y_true` and `y_predict`.
- Assign the performance score to the `score` variable.

```python
# TODO: Import 'r2_score'
from sklearn.metrics import r2_score

def performance_metric(y_true, y_predict):
    """ Calculates and returns the performance score between 
        true and predicted values based on the metric chosen. """
    
    # TODO: Calculate the performance score between 'y_true' and 'y_predict'
    score = r2_score(y_true, y_predict)
    
    # Return the score
    return score
```

**Answer: **
A <b>técnica de busca em grade</b> (grid search technique) se consiste em variar os parâmetros de um algoritmo de aprendizagem de maneira parametrizada, treinar o modelo e testar nos dados de validação, anotando seu desempenho (através de alguma técnica de avaliação como <b>r2_score, f1_score, curva ROC,</b> etc). O desempenho de cada variação de parâmetros é organizada através de uma tabela, indicando qual foi o desempenho final para cada par de parâmetros.

Por exemplo, caso um algoritmo de aprendizagem tenha os parâmetros <i>X e Y</i>, podemos organizá-los da seguinte maneira:

|ParâmetroX / ParâmetroY| X1 | X2 | X3 | 
|:-------------:| :-------------: | :--------: |
|Y1| F1_Score: 0.5 | F1_Score: 0.2 | F1_Score: 0.7 | 
|Y2| F1_Score: 0.4 | F1_Score: 0.9 | F1_Score: 0.6 | 
|Y3| F1_Score: 0.1 | F1_Score: 0.5 | F1_Score: 0.7 | 
|Y4| F1_Score: 0.2 | F1_Score: 0.7 | F1_Score: 0.2 | 

Onde <i>X1, X2, X3, Y1, Y2, Y3</i> e <i>Y4</i> são os valores dos parâmetros organizados de maneira estruturada (por exemplo assumindo valores como 0, 1, 10, 100, 1000, etc). Para cada par de valores de parâmetros utilizamos alguma técnica de avaliação (F1_Score no exemplo) para avaliar o modelo. Nesse exemplo, o algoritmo com os valores <b>X = X2</b> e <b>Y = Y2</b> resultaram no melhor F1_Score (0.9). Isso indica que, dentre as possibilidades testadas, esse par de parâmetros fornece o melhor resultado do modelo. 

Esse processo fornece uma maneira estruturada de ajustar os parâmetros de um algoritmo de aprendizado, auxiliando a decidir os melhores parâmetros para o modelo.

A validação k-fold particiona os dados em k segmentos, por exemplo, um k-fold com k = 4 particiona os dados em 4 conjuntos diferentes. A partir desta partição de k segmentos, o modelo então deve ser treinado k vezes, cada vez utilizando um conjunto diferente como dados de teste e os k-1 conjuntos restantes como os dados de treino. A média dos resultados de cada conjunto geram o score final daquele modelo.

A técnica de busca em grade pode se beneficiar do k-fold através do treinamento do modelo em diversos grupos de dados diferentes. Desta forma, caso os dados originais estejam organizados de alguma maneira mais vantajosa para um par de parâmetros específico, o método k-fold pode reorganizar os dados e comprovar que o resultado daquele conjunto de parâmetros é válido para várias variações de conjuntos de dados, evitando desperdício de dados e obtendo uma avaliação melhor do modelo.

Desta forma, para cada par de parâmetros da busca em grade, o modelo será treinado e testado através da validação k-fold, e o resultado final de cada conjunto de parâmetros será baseado na média dos resultados da validação k-fold.

```python
# TODO: Import 'make_scorer', 'DecisionTreeRegressor', and 'GridSearchCV'
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer, f1_score

def fit_model(X, y):
    """ Performs grid search over the 'max_depth' parameter for a 
        decision tree regressor trained on the input data [X, y]. """

    # Create cross-validation sets from the training data
    # sklearn version 0.18: ShuffleSplit(n_splits=10, test_size=0.1, train_size=None, random_state=None)
    # sklearn versiin 0.17: ShuffleSplit(n, n_iter=10, test_size=0.1, train_size=None, random_state=None)
    cv_sets = ShuffleSplit(X.shape[0], n_iter = 10, test_size = 0.20, random_state = 0)

    # TODO: Create a decision tree regressor object
    regressor = DecisionTreeRegressor()

    # TODO: Create a dictionary for the parameter 'max_depth' with a range from 1 to 10
    params = {'max_depth': range(1,11)}
    print params

    # TODO: Transform 'performance_metric' into a scoring function using 'make_scorer' 
    scoring_fnc =  make_scorer(performance_metric)

    # TODO: Create the grid search cv object --> GridSearchCV()
    # Make sure to include the right parameters in the object:
    # (estimator, param_grid, scoring, cv) which have values 'regressor', 'params', 'scoring_fnc', and 'cv_sets' respectively.
    grid = GridSearchCV(regressor, params, scoring_fnc, cv=cv_sets)

    # Fit the grid search object to the data to compute the optimal model
    grid = grid.fit(X, y)

    # Return the optimal model after fitting the data
    return grid.best_estimator_
```
